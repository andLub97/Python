```python
vendas = [1500, 2000, 4000,50]
meses = ['Jan','Fev','Mar','Abr']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(meses, vendas)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.axis([0, 3, 0,5000])
plt.show()
```

```python
import numpy as np
```

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
```

```python
plt.plot(meses,vendas)
plt.axis([0, 50, 0, max(vendas)+200])
plt.show()
```

```python
plt.plot(meses,vendas,'c--')
plt.axis([0, 50, 0, max(vendas)+200])
plt.show()
```

```python
plt.scatter(meses,vendas)
plt.show()
```

```python
plt.bar(meses,vendas)
plt.show()
```
